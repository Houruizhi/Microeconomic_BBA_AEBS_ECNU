```python
import xlrd
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
path = 'Anwser_Mini-Course_in_Crypto_Economics_After-course_questionaire.xlsx'
```

```python
def load_survey(path):
    survey = xlrd.open_workbook(path)
    table = survey.sheet_by_name('Sheet1')
    return table
```

```python
table = load_survey(path)
col_values = table.col_values(1)
print(col_values)
title_name = col_values[0]
record_number = len(col_values) - 1
```

```python
def count_num(data):
    value_vs_number = {}
    for i in data:
        i = i.split(',')[0]
        if i in value_vs_number.keys():
            value_vs_number[i] += 1
        else:
            value_vs_number[i] = 1
    return value_vs_number
```

```python
def load_data(path, colume):
    table = load_survey(path)
    col_values = table.col_values(colume)
    title = col_values[0].split("、")[-1]
    col_values = col_values[1:]
    return col_values, title
def draw_figure(data, order=None, palette = ['red','blue']):
    xx = list(data.values())
    index = np.argsort(xx)
    xx.sort(reverse=True)
    yy = [list(data.keys())[i] for i in index[::-1]]
    ax = sns.barplot(x=xx,y=yy, order=order, orient='h', palette=palette)
    ax.set_xlabel("Number of Students", fontsize=14)
    if order==None:
        patches = ax.patches
    else:
        patches = ax.patches[:len(order)]
    total = 0
    for i in patches:
        total += i.get_width()
    ax.set_xticks(list(range(0,int(ax.patches[0].get_width())+21,5)))
    for i in patches:
        ax.text(i.get_width(), i.get_y()+0.5, f'{str(int(i.get_width()))}({str(round((i.get_width()/total)*100, 2))}%)', fontsize=12, color='gray')
    return ax
```

```python
def answer_to_txt(col_values, save_path):
    with open(save_path, 'w', newline='') as f:
        for i in col_values:
            if 'Yes' in i:
                f.write(i.split('〖')[-1].split('〗')[0]+'\n')
```

```python
plt.figure(dpi=100)
col_values, title = load_data(path,7)
data = count_num(col_values)
draw_figure(data)
plt.title('Gender')
save_path = './pictures/gender.png'
plt.savefig(save_path,bbox_inches='tight')
```

```python
stu_yes = {}
for i in range(1, table.nrows):
    row = table.row_values(i)
    std_id = row[8]
    yes_num = 0
    for k in range(8, table.ncols):
        if 'Yes' in row[k]:
            yes_num += 1
    stu_yes[std_id] = yes_num
```

```python
stu_yes_sorted = sorted(stu_yes.items(), key=lambda item:item[1], reverse=True)
```

```python
stu_id_list = table.col_values(8)
```

```python
import csv
with open('yes_num.csv','w',newline='') as f:
    wrtier = csv.writer(f)
    wrtier.writerow(['Last Name','First Name','ECNU Student ID','Yes number'])
    for stu_id, num in stu_yes_sorted:
        if stu_id in stu_id_list:
            index = stu_id_list.index(stu_id)
            row = table.row_values(index)
            wrtier.writerow([row[6],row[7],row[8],num])
```
